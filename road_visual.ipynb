```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sf_edge_pos.csv")
display(df)
```

```python
for index, row in df.iterrows():
    plt.plot([row['start_x'], row['end_x']], [row['start_y'], row['end_y']])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Lines from DataFrame')
plt.grid(True)
plt.show()
```

```python
node_dist = pd.read_csv('node_distances.csv')
plt.figure(figsize=(8, 5))  # Set the figure size
cmap = 'viridis'

# Plot the scatter plot with color grading based on distance values
scatter = plt.scatter(node_dist['node_x'], node_dist['node_y'], c=node_dist['average_distance'], cmap=cmap, s=100, alpha=0.8, label='Scatter Plot')
plt.colorbar(scatter, label='Distances')

plt.show()
```

```python
#Graph of Dijkstra's weighted shortest distances
node_dist = pd.read_csv('node_weighted_distances.csv')
plt.figure(figsize=(8, 5))  # Set the figure size
cmap = 'viridis'

# Plot the scatter plot with color grading based on distance values
scatter = plt.scatter(node_dist['node_x'], node_dist['node_y'], c=node_dist['average_distance'], cmap=cmap, s=100, alpha=0.8, label='Scatter Plot')
plt.colorbar(scatter, label='Weights')

plt.show()
```
